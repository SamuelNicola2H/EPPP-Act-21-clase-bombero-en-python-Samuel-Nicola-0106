```python
class Bombero:
    # El método __init__ es el constructor: inicializa los atributos
    def __init__(self, nombre, rango, estacion):
        self.nombre = nombre
        self.rango = rango
        self.estacion = estacion

    # Definición de las funciones (métodos)
    def salvar(self):
        return f"{self.nombre} está rescatando a una persona en peligro."

    def apagar_incendios(self):
        return f"{self.nombre} ha conectado la manguera y está extinguiendo el fuego."

    def atencion_medica(self):
        return f"{self.nombre} está brindando primeros auxilios básicos."

# --- Creación de Objetos (Instancias) ---

# Bombero 1
bombero1 = Bombero("Marcos Ruiz", "Capitán", "Estación Central")

# Bombero 2
bombero2 = Bombero("Elena Gómez", "Teniente", "Estación Sur")

# --- Uso de los objetos y sus mensajes ---

print(f"--- Ficha de {bombero1.nombre} ---")
print(f"Rango: {bombero1.rango}")
print(f"Estación: {bombero1.estacion}")
print(f"Acción: {bombero1.salvar()}")

print("\n" + "-"*30 + "\n")

print(f"--- Ficha de {bombero2.nombre} ---")
print(f"Rango: {bombero2.rango}")
print(f"Estación: {bombero2.estacion}")
print(f"Acción: {bombero2.apagar_incendios()}")

print("\n" + "="*30)
print("programa creado por Samuel Nicola NL = 40")
```
